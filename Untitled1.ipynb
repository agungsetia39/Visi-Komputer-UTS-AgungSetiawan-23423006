```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab import files
```

```python
uploaded = files.upload()

# Ambil nama file
image_path = list(uploaded.keys())[0]

# Baca gambar
img = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title("Citra Asli")
plt.axis('off')
```

```python
# Resize (opsional)
img_resized = cv2.resize(img_rgb, (300, 300))

# Grayscale
gray = cv2.cvtColor(img_resized, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.title("Grayscale")
plt.axis('off')
```

```python
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
sobel = np.sqrt(sobelx**2 + sobely**2)

plt.imshow(sobel, cmap='gray')
plt.title("Sobel")
plt.axis('off')
```

```python
kernelx = np.array([[1,0,-1],[1,0,-1],[1,0,-1]])
kernely = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])

prewittx = cv2.filter2D(gray, -1, kernelx)
prewitty = cv2.filter2D(gray, -1, kernely)
prewitt = np.sqrt(prewittx**2 + prewitty**2)

plt.imshow(prewitt, cmap='gray')
plt.title("Prewitt")
plt.axis('off')
```

```python
canny = cv2.Canny(gray, 100, 200)

plt.imshow(canny, cmap='gray')
plt.title("Canny")
plt.axis('off')
```

```python
kernel = np.ones((3,3), np.uint8)

dilation = cv2.dilate(gray, kernel)
erosion = cv2.erode(gray, kernel)

morph_gradient = dilation - erosion

plt.imshow(morph_gradient, cmap='gray')
plt.title("Morfologi Gradien")
plt.axis('off')
```

```python
titles = ['Original', 'Grayscale', 'Prewitt', 'Sobel', 'Canny', 'Morph Gradient']
images = [img_resized, gray, prewitt, sobel, canny, morph_gradient]

plt.figure(figsize=(12,8))
for i in range(6):
    plt.subplot(2,3,i+1)
    if i == 0:
        plt.imshow(images[i])
    else:
        plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.show()
```
